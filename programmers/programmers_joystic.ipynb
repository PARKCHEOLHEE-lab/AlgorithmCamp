```python
name = "JAACAAN"
def solution(name):
    
    if set(name) == {"A"}:
        return 0
    
    abNum = {'A':0}
    asc = ['B','C','D','E','F','G','H','I','J','K','L','M','N']
    dsc = ['Z','Y','X','W','V','U','T','S','R','Q','P','O','N']
    
    for i in range(len(asc)):
        abNum[asc[i]] = i+1
        abNum[dsc[i]] = i+1
        
    right = abNum[name[0]]
    for i in range(1, len(name)):
        right += abNum[name[i]]+1
    
    a_len = 0
    for i in range(1, len(name)):
        if 'A'*i in name:
            a_len = i    
            
    a_idx = name.find(a_len*'A')
    left = abNum[name[0]]
    for i in range(1, a_idx):
        left += abNum[name[i]]+2
    
    for i in range(-1, -(len(name) - (a_idx+a_len)+1), -1):
        left += abNum[name[i]]+1
    
    return min(left, right)

solution(name)
```
