### quiz1 : 프로그래머스 짝지어 제거하기
> 

```python
S = "baabaa"
s = "baabaa"

s.replace("aa", "")
```

```python
for i in range(len(S)-1):
    if S[i] == S[i+1]:
        ds = S[i]*2
        s = s.replace(ds, "")
        
s
```

```python
s = "baabaa"

def solution(s):
    stack = []
    for i in range(len(s)):
        if len(stack) == 0:
            stack.append(s[i])
        else:
            if s[i] == stack[-1]:
                stack.pop()
            else:
                stack.append(s[i])

    if len(stack) != 0:
        return 0
    else:
        return 1
        
print(solution(s))
```

### quiz2 : leetcode REVERSE INTEGER
>

```python
class Solution:
    def reverse(self, x: int) -> int:
        X = str(abs(x))
        
        if (not -2**31 <= int(X[::-1]) <= 2**31-1) or (not -2**31 <= x <= 2**31-1):
            return 0
                
        elif x < 0:
            return -int(X[::-1])
            
        else:
            return int(X[::-1])
        
print(Solution().reverse(-2**31))
print(Solution().reverse(2**31))
print(Solution().reverse(-123))
print(Solution().reverse(0))
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        X = str(abs(x))
        revNum = int("".join([X[i] for i in range(len(X)-1,-1,-1)]))
        
        if (not -2**31 <= revNum <= 2**31-1) or (not -2**31 <= x <= 2**31-1):
            return 0
                
        else: 
            return -revNum if x < 0 else revNum
            
print(Solution().reverse(-2**31))
print(Solution().reverse(2**31))
print(Solution().reverse(-123))
print(Solution().reverse(120))
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        C = 2**31
        X = str(abs(x))
        revNum = int("".join([X[i] for i in range(len(X)-1,-1,-1)]))
        
        if (not -C <= revNum <= C-1) or (not -C <= x <= C-1):
            return 0
                
        else: 
            return -revNum if x < 0 else revNum
            
print(Solution().reverse(-2**31))
print(Solution().reverse(2**31))
print(Solution().reverse(-123))
print(Solution().reverse(120))
```

### quiz3 : leetcode Third Maximum Number
>

```python
from typing import List

nums = [1,2,3,5,7,89,9,1,1]
class Solution:
    def thirdMax(self, nums: List[int]) -> int:
        nums = list(set(nums))
        nums.sort(reverse=True)
        
        return nums[2] if len(nums) >= 3 else max(nums)

Solution().thirdMax(nums)
```

### quiz4 : leetcode Add Two Numbers
>

```python
l1 = [0]
l2 = [0]

class Solution:
    def revNum(li):
        revList = []
        for i in range(len(li)-1, -1, -1):
            revList.append(str(li[i]))

        return int("".join(revList))
    
    def addNumRev(li1, li2):
        addNum = str(revNum(li1) + revNum(li2))
        revList = []
        for i in range(len(addNum)-1, -1, -1):
            revList.append(int(addNum[i]))
        
        return revList

Solution.addNumRev(l1, l2)
```

### quiz6 : leetcode Plus One
> 

```python
digits = [9]

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        
        L = len(digits)-1
        
        for i in range(L, -1, -1):
            
            if digits[i] != 9:
                digits[i] = digits[i]+1
                break
                
            else:
                digits[i] = 0
                carry = 1
        
        if digits[0] == 0 and carry == 1:
            digits.insert(0,1)
            
        return digits
                        
Solution().plusOne(digits)
```

### quiz7: leetcode Happy Number
> 

```python
n = 2

class Solution:
    def isHappy(self, n: int) -> bool:
        s = []
        while n not in s:
            s.append(n)
            n = sum(int(i)**2 for i in str(n))
            if n == 1:
                return True
        return False
            
Solution().isHappy(n)
```

### quiz8 : leetcode Single Number
> 

```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        for i in nums:
            if nums.count(i) == 1:
                return i
                break
```

### quiz9 : leetcode Add Digits
>

```python
class Solution:
    def addDigits(self, num: int) -> int:
        while len(str(num)) > 1:
            num = list(str(num))
            num = sum(list(map(int, num)))

        return num
```

### quiz10 : leetcode Majority Element
> 

```python
from typing import List

nums = [2,2,1,1,1,2,2,1,1,1,5,5,5,5,5,5,5,5,5]

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        sNum = list(set(nums))
        sDict = {}
        
        for i in sNum:
            sDict[i] = nums.count(i)
        
        return  sorted(sDict.items(), key=lambda x : x[1])[-1][0]
        
Solution().majorityElement(nums)
```
