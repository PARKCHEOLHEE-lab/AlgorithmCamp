### quiz8 : 프로그래머스 전화번호 목록
> 

```python
p = ["12","13","14","15","1"]

def solution(p):
    p.sort()
    for i in range(len(p)-1):
        if len(p[i]) < len(p[i+1]):
            if p[i+1][:len(p[i])] == p[i]:
                return False
                break
    return True    

solution(p)
```
