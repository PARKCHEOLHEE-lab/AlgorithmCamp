### quiz2 : 프로그래머스 평균 구하기
> 

```python
# 넘파이 라이브러리 사용

import numpy as np

arr = [1,2,3,4]

def solution(arr):
    return np.mean(arr)

print(solution(arr))
```

### quiz3 : 프로그래머스 행렬의 덧셈
> 

```python
arr1 = [[1,2], [2,3]]
arr2 = [[3,4], [5,6]]

def solution(arr1, arr2):    
    answer = arr1
    
    for i in range(len(arr1)):
        for j in range(len(arr1[i])):
            answer[i][j] = arr1[i][j] + arr2[i][j]

    return answer

solution(arr1, arr2)
```

### quiz4 : 프로그래머스 짝수와 홀수

### quiz5 : 프로그래머스 자릿수 더하기
> 

```python
n = 987

def solution(n):
    n = str(n)
    answer = 0
    
    for i in n: answer += int(i)
    return answer

solution(n)
```

### quiz6 : 프로그래머스 최대공약수와 최소공배수
> 

```python
from math import gcd

n = 6
m = 12

def solution(n, m):
    
    return [gcd(n,m), int(n*m/gcd(n,m))]

solution(n, m)
```

### quiz7 : 프로그래머스 정수 제곱근 판별
> 

```python
import math

n = 121

def solution(n):
    
    if n % math.sqrt(n) == 0:
        return (math.sqrt(n)+1)**2
    
    else:
        return -1
    
solution(n)
```

```python
n = 3

def solution(n):
    if n % n**(1/2) == 0:
        return (n**(1/2)+1)**2
    
    else:
        return -1
    
solution(n)
```

### quiz8 : 프로그래머스 문자열 다루기 기본
> 

```python
s = "a12345"

def solution(s):
    try:
        if len(s) == 4 or len(s) == 6:
            s = int(s)
            return True
        
        else:
            return False
        
    except:
        return False

solution(s)
```

### quiz9 : 프로그래머스 정수 내림차순으로 배치하기
> 

```python
n = 118372

def solution(n):
    n = str(n)
    n = list(map(int, n))
    n.sort(reverse=True)
    n = list(map(str, n))
    
    return int("".join(n))
    
solution(n)
```

### quiz10 : 프로그래머스 자연수 뒤집어 배열로 만들기
> 

```python
n = 1007007888

def solution(n):
    return [int(str(n)[i]) for i in range(len(str(n))-1, -1, -1)]

solution(n)
```
