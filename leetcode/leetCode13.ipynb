```python
class Solution:
    def romanToInt(self, s: str) -> int:
        
        roman = {
                 "I"  : 1,
                 "V"  : 5, 
                 "X"  : 10, 
                 "L"  : 50,
                 "C"  : 100,
                 "D"  : 500,
                 "M"  : 1000
                }
        
        answer = 0        
        for i in range(len(s)-1):        
            curr_num = roman[s[i]]
            next_num = roman[s[i+1]]
            
            if curr_num < next_num:
                answer -= curr_num
            
            else:
                answer += curr_num
                
        answer += roman[s[-1]]
        
        return answer
           
s = "MCMXCIV"
Solution().romanToInt(s)
```
