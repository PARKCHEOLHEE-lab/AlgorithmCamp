```python
from typing import List

nums = [0,0,1,1,1,2,2,3,3,4]
nums = [1,1,2]
nums = [-1,0,0,0,0,3,3]

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        
        check = nums[0]
        idx = 0        
        for n in nums:
            if check != n:
                idx = idx+1
                nums[idx] = n
                check = nums[idx]
        
        return len(nums[:len(set(nums))])

print(Solution().removeDuplicates(nums))
```
